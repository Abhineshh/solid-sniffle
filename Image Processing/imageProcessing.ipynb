### Resizing the image 

```python
from PIL import Image
```

```python
img = Image.open("casey_stoner.jpg")
img_resized = img.resize((1920,1080))
```

```python
img_resized.save("casey_stone_1080p.jpg") # saving the resized image
```

### Converting RGB to GrayScale

```python
import cv2 # import opencv
```

```python
pic = cv2.imread("casey_stoner.jpg")
```

```python
type(pic)
```

```python
pic.shape
```

```python
grayscale_img = cv2.cvtColor(pic, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```

```python
# cv2.imshow(image_name) will display the image but it works in a file environment and 
# would not work in a notebook environment
gray_img_plot = plt.imshow(grayscale_img)
plt.show()
```

```python
cv2.imwrite('grayscale_casey_stone.jpg', grayscale_img)
```
