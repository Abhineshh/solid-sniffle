# Styling

```python
x = [1,2,3,4]
dsa_marks = [90,92,96,98]
ml_marks = [93,95,98,100]

plt.plot(x,dsa_marks, label= 'dsa_marks',color='red', linewidth='4', linestyle="--")
plt.plot(x, ml_marks, label = 'ml_marks', color='#34ae23', marker='*', markersize = 10)

plt.title("Marks corresponding to no of hrs studied",
         fontdict={'fontname':'times new roman', 'fontweight': 10, 'fontsize': 30, 'color': 'blue'}
         )
plt.xlabel("No of hrs studied",
          fontdict={'fontname':'times new roman', 'fontweight': 10, 'color': 'green'}
      )
plt.ylabel("Marks",
           fontdict={'fontname':'times new roman', 'fontweight': 10, 'color': 'green'}
      )

plt.xticks([1,2.5,4]) # specifying the x axis values labelling
plt.yticks([90,95,100]) # specifying the y axis values labelling

plt.legend(loc='upper left') # a sticker on the chart specifying the colors indication
# plt.legend(loc='upper right')
# plt.legend(loc='lower left')
# plt.legend(loc='lower right')

plt.show()
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)
print(t)
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

# plots in multiple figures

```python
t = np.arange(0., 5., 0.2)

plt.figure()
plt.plot(t, t, 'r--')
plt.show()

plt.figure()
plt.plot(t, t**2, 'bs')
plt.show()

plt.figure()
plt.plot(t, t**3, 'g^')
plt.show()
```

# subplots

```python
t = np.arange(0., 5., 0.2)
# horizontal placement of plots
plt.figure()
plt.subplot(1,3,1) # [row, column, current plot index]
plt.plot(t, t, 'r--')

plt.subplot(1,3,2) # [row, column, current plot index]
plt.plot(t, t**2, 'bs')

plt.subplot(1,3,3) # [row, column, current plot index]
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
# vertical stacking of plots
plt.figure()
plt.subplot(3,1,1) # [row, column, current plot index]
plt.plot(t, t, 'r--')

plt.subplot(3,1,2) # [row, column, current plot index]
plt.plot(t, t**2, 'bs')

plt.subplot(3,1,3) # [row, column, current plot index]
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
t = np.arange(0., 5., 0.2)

plt.figure()
plt.subplot(2,3,1) # [row, column, current plot index]
plt.plot(t, t, 'r--')

plt.subplot(2,3,2) # [row, column, current plot index]
plt.plot(t, t**2, 'bs')

plt.subplot(2,3,3) # [row, column, current plot index]
plt.plot(t, t**3, 'g^')

plt.subplot(2,3,4) # [row, column, current plot index]
plt.plot(t, t**2, 'bs')

plt.subplot(2,3,5) # [row, column, current plot index]
plt.plot(t, t, 'r--')

plt.subplot(2,3,6) # [row, column, current plot index]
plt.plot(t, t**3, 'g^')
plt.savefig("ding.png")
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
# horizontal placement of plots
plt.figure()

plt.subplot(1,3,1) # [row, column, current plot index]
plt.plot(t, t, 'r--')
plt.title("linear subplot")
plt.xlabel("x axis points")
plt.ylabel("y axis points")

plt.subplot(1,3,2) # [row, column, current plot index]
plt.plot(t, t**2, 'bs')
plt.title("quadratic subplot")
plt.xlabel("x axis points")
plt.ylabel("y axis points")

plt.subplot(1,3,3) # [row, column, current plot index]
plt.plot(t, t**3, 'g^')
plt.title("Cubic subplot")
plt.xlabel("x axis points")
plt.ylabel("y axis points")

plt.suptitle("subplot of the century")

plt.tight_layout() # adjust the spacing
plt.show()
```

# Subplotting with Axes

```python
x = np.linspace(1,10,100) # here start, end, no. of values needed(evenly distributed)

fig, axes = plt.subplots(2,3) # telling to create 6 subplots in 2 rows and 3 columns

axes[0,0].plot(x, np.sin(x))
axes[0,0].set_title('Sin Function')
axes[0,0].set_xlabel("x")
axes[0,0].set_ylabel('sinx = y')

axes[1,0].plot(x, np.cos(x))
axes[1,0].set_title('cos Function')
axes[1,0].set_xlabel("x")
axes[1,0].set_ylabel('cosx = y')

axes[0,1].plot(x, np.tan(x))
axes[0,1].set_title('tan Function')
axes[0,1].set_xlabel("x")
axes[0,1].set_ylabel('tanx = y')
                     
                     
axes[1,1].plot(x, (1/np.tan(x)))
axes[1,1].set_title('cot Function')
axes[1,1].set_xlabel("x")
axes[1,1].set_ylabel('cotx = y')
                     
axes[0,2].plot(x, (1/np.sin(x)))
axes[0,2].set_title('cosec Function')
axes[0,2].set_xlabel("x")
axes[0,2].set_ylabel('cosecx = y')
                     
axes[1,2].plot(x, (1/np.cos(x)))
axes[1,2].set_title('sec Function')
axes[1,2].set_xlabel("x")
axes[1,2].set_ylabel('secx = y')


plt.tight_layout()
```

# Scatter Plots

```python
xrr = np.random.rand(500) * 10 # between 0 and 10
yrr = np.random.rand(500) * 10
print(xrr, yrr)
```

```python
plt.scatter(xrr,yrr, marker='s',label="random", alpha=0.5, s = 50) # p mean pentagon
# here alpha specifie transparency between 0 and 1 to show overlapping in points
# s is size

plt.legend(loc="lower right")

plt.grid(True)

plt.show()
```

# Bar Charts

```python
categories = ['physics','chemistry','Mathematics']
marks = [63,61,65]

plt.bar(categories, marks, )

plt.title("mark Graph")
plt.xlabel("subjects")
plt.ylabel("marks")

plt.show()
```

```python
categories = ['physics','chemistry','Mathematics']
marks = [63,61,65]
bar_color = ['blue','green','red']
bars = plt.bar(categories, marks, )

for bar,color in zip(bars, bar_color):
    bar.set_color(color)

plt.xlabel("subjects")
plt.ylabel("marks")

plt.show()
```
