# Accessing Data

## loc: 
- loc helps to list multiple columen and rows at a time
- it support column name bases indexing/labelling also
- it also support boolean based indexing

```python
print(obesity_data.loc[0]) #getting a particular row by index/label
print(obesity_data.loc[0,'Age']) # only a particular column of a specific row
print(obesity_data.loc[0:10, 'Age']) # both start and end are inclusive
print(obesity_data.loc[0:10, ['Age','Height','Weight','Gender']]) # multiple comlumns 
print(obesity_data.loc[[1,4,3,5,24],['Age','Height','Weight','Gender']]) # getting only rows you want 
print(obesity_data.loc[0:5, 'FAVC':'TUE' ]) # getting columns by index
```

## iloc: 
- when we use slicing in iloc the last index is excluded unlike loc
- in iloc we use index labelling meaning normal integer index, unlike loc where we can use column name(column label)

```python
print(obesity_data.iloc[0:5,0]) # index '5' is excluded
#  print(obesity_data.iloc[0:5,'Age'] # give error becouse column name labelling is not allowed in iloc
print(obesity_data.iloc[0:5,0:5])
print(obesity_data.loc[[1,4,3,5,24],:])
```

## Single Value Access:  at , iat
- slicing is not allowed in at and iat.
- at is similar to loc and iat to iloc, but slicing is not allowed
- even lists to access specific rows or columns is not allowed
- in loc and iloc column label is not mandatory but its mandatory in at and iat

```python
print(obesity_data.at[0,'Age'])
print(obesity_data.at[4,'Height'])
#   print(obesity_data.iat[0]) # give error
#   print(obesity_data.at[0]) # give error
print(obesity_data.iat[0,2])
```

# Transformations
## Apply() give a new dataframe with the applied changes and does not change the original dataframe

```python
def increment_by_5(x):
    return x + 5
def decrement_by_5(x):
    return x - 5


obesity_data['Age'] = obesity_data['Age'].apply(increment_by_5)
print(obesity_data['Age'])
obesity_data['Age'] = obesity_data['Age'].apply(decrement_by_5)
print(obesity_data['Age'])
```

### lambda function with apply()

```python
obesity_data['Age'] = obesity_data['Age'].apply(lambda x: x - 5)
print(obesity_data['Age'])
```

# Merging DataFrames

## merging dataframes is similar to joins in sql
### inner join
### outer join
### left join
### right join

```python
df1 = pd.DataFrame(
    {
        "id": [1,2,3,4],
        "name": ["Abhinesh","KrishnaPriya","A","K"]
    }
)
df2 = pd.DataFrame(
    {
        "id": [3,4,5,6],
        "skill": ["Scientist","Artist","Engineer","Teacher"]
    }
)
inner_join = pd.merge(df1,df2, how='inner', on="id") #on is the primary key
print(inner_join)
outer_join = pd.merge(df1,df2, how='outer', on="id")
print(outer_join)
left_join = pd.merge(df1,df2, how='left', on="id")
print(left_join)
right_join = pd.merge(df1,df2, how='right', on="id")
print(right_join)
```

## Sometime the columnname  maybe different so,


```python
df1 = pd.DataFrame(
    {
        "id1": [1,2,3,4],
        "name": ["Abhinesh","KrishnaPriya","A","K"]
    }
)
df2 = pd.DataFrame(
    {
        "id2": [3,4,5,6],
        "skill": ["Scientist","Artist","Engineer","Teacher"]
    }
)
inner_join = pd.merge(df1,df2, how='inner', left_on="id1", right_on="id2") # left_on is df1's primary key and right_on is df2's primary key
print(inner_join)
outer_join = pd.merge(df1,df2, how='outer', left_on="id1", right_on="id2")
print(outer_join)
left_join = pd.merge(df1,df2, how='left', left_on="id1", right_on="id2")
print(left_join)
right_join = pd.merge(df1,df2, how='right', left_on="id1", right_on="id2")
print(right_join)
```

# Concatenating DataFrams
### its not merging where we need primary key, it just concatenates by vertically stacking or fusing horizontally

```python
concated_df = pd.concat([df1, df2], axis=0)
# axis=0 is vertical Stacking while 1 is horizontal
print(concated_df)
```

```python
# since it is taking previous index, to assign a new index
concated_df = pd.concat([df1,df2], axis = 0, ignore_index = True)
print(concated_df)
```

```python
# horizontal merge
concated_hdf = pd.concat([df1,df2], axis = 1)
print(concated_hdf)
```

```python
# multilevel indexing
concated_df = pd.concat([df1,df2], axis = 0, keys=["dataF1", "dataF2"])
print(concated_df)
```
